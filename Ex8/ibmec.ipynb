```python
import requests
import pandas as pd
import os

token = "SEU_TOKEN_AQUI"

headers = {
    "Authorization": f"Bearer {token}",
    "Accept": "application/json",
    "User-Agent": "Mozilla/5.0",
    "Referer": "https://estudante.ibmec.br/",
    "Origin": "https://estudante.ibmec.br"
}

turmas_ids = [
    "ibmec_14288731",
    "ibmec_14288775",
    "ibmec_14288583",
    "ibmec_14288612",
    "ibmec_14288690"
]

os.makedirs("arquivos", exist_ok=True)

dados = []

for turma_id in turmas_ids:

    url_turma = f"https://apis.estudante.ibmec.br/rest/turmas/{turma_id}?matricula=202407096301"
    r_turma = requests.get(url_turma, headers=headers)

    if r_turma.status_code != 200:
        continue

    nome_disciplina = r_turma.json()["disciplina"]["nome"]

    url_conteudos = f"https://apis.estudante.ibmec.br/rest/v2/entregas/{turma_id}/conteudos-complementares"
    r_cont = requests.get(url_conteudos, headers=headers)

    if r_cont.status_code == 200:
        conteudos = r_cont.json()

        for i, item in enumerate(conteudos):

            url = item.get("url")
            caminho_local = None

            if url:
                try:
                    r_file = requests.get(url)

                    content_type = r_file.headers.get("Content-Type", "")

                    # 🔥 define extensão pelo content-type
                    if "pdf" in content_type:
                        ext = "pdf"
                    elif "zip" in content_type:
                        ext = "zip"
                    elif "rar" in content_type:
                        ext = "rar"
                    elif "msword" in content_type:
                        ext = "doc"
                    else:
                        print(f"⚠️ Tipo desconhecido: {content_type}")
                        continue  # NÃO salva lixo

                    if r_file.status_code == 200 and "html" not in content_type:
                        nome = f"{turma_id}_{i}.{ext}"
                        caminho_local = f"arquivos/{nome}"

                        with open(caminho_local, "wb") as f:
                            f.write(r_file.content)

                        print(f"✅ Baixado: {nome}")
                    else:
                        print(f"❌ Conteúdo inválido: {item.get('titulo')}")

                except Exception as e:
                    print(f"Erro: {e}")

            dados.append({
                "turma_id": turma_id,
                "disciplina": nome_disciplina,
                "titulo": item.get("titulo"),
                "tipo": item.get("tipo"),
                "url": caminho_local if caminho_local else url
            })

df = pd.DataFrame(dados)
df.to_csv("conteudos_completos.csv", index=False)

print("\n🔥 Agora sim, sem .bin lixo!")
```

```python
def classificar_tipo(row):
    tipo = row["tipo"]
    url = str(row["url"]).lower()

    if tipo == "arquivo":
        if url.endswith(".pdf"):
            return "PDF"
        elif url.endswith(".doc") or url.endswith(".docx"):
            return "DOC/DOCX"
        else:
            return "IPYNB"

    elif tipo == "link":
        if "colab.research.google.com" in url:
            return "GOOGLE_COLAB"
        elif "docs.google.com" in url:
            return "GOOGLE_DOCS"
        elif "drive.google.com" in url:
            return "GOOGLE_DRIVE"
        else:
            return "OUTRO_LINK"

    return "DESCONHECIDO"
```
